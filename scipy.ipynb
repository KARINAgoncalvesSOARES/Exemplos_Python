> 1. Integração Numérica:

O SciPy fornece funções para realizar integração numérica, como quad para integração de funções simples e dblquad para integração de funções bidimensionais.

```python
from scipy import integrate
from scipy.integrate import quad
```

```python
x2 = lambda x: x**2
integral = quad(x2, 0, 2)
print('resultado: {}, erro: {}'.format(integral[0], integral[1]))
```

```python
result, error = integrate.quad(lambda x: x**2, 0, 1)
print('resultado: {}, erro: {}'.format(result, error))
```

```python
import numpy as np

def normal_fdp(x, mu, sigma_2):
    return np.exp(-(x - mu) ** 2 / (2 * sigma_2)) / np.sqrt(2 * np.pi * sigma_2)

standard_fdp = lambda x: normal_fdp(x, 0, 1)
integral = quad(standard_fdp, -3, 3)
print('P(-3 < X < 3): {}, erro: {}'.format(integral[0], integral[1]))
```

Para computar integrais duplas, triplas e sobre múltiplas variáveis, o módulo oferece respectivamente as funções dblquad, tplquad e nquad.



> 2. Otimização:

Você pode usar o módulo optimize para resolver problemas de otimização, como minimizar ou maximizar funções.

```python
from scipy.optimize import minimize
result = minimize(lambda x: x**2, x0=0)
print('Valor mínimo encontrado:', result.fun)
```

> 3. Álgebra Linear:

O SciPy inclui uma série de funções para lidar com problemas de álgebra linear, como resolução de sistemas de equações lineares e cálculo de autovalores e autovetores.

```python
from scipy import linalg
A = np.array([[1, 2], [3, 4]])
b = np.array([5, 6])
x = linalg.solve(A, b)
print(x)
```

> 4. Processamento de Sinais:

O módulo signal do SciPy oferece ferramentas para processamento de sinais, como filtragem, transformada de Fourier e análise espectral.

```python
from scipy import signal
fs = 1000  # Frequência de amostragem
t = np.linspace(0, 1, fs, endpoint=False)  # Vetor de tempo
signal_data = np.sin(2 * np.pi * 5 * t)  # Sinal senoidal
filtered_signal = signal.lfilter([1.0], [1.0, -0.5], signal_data)
print(filtered_signal)
```

> 5. Estatísticas:
O SciPy oferece funções estatísticas para cálculos como teste t, correlação, ajuste de distribuições, etc.

```python
from scipy import stats
data = np.random.normal(0, 1, 100)  # Amostra de dados
mean, p_value = stats.ttest_1samp(data, 0)
print('Média da amostra:', mean)
print('Valor p:', p_value)
```
